```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
import struct
import numpy as np

with open("C:\\Users\\tymot\\projects\\001.smartula\\smartula-analysis\\smartula-analysis\\measurements\\smartulav2\\ul6\\270073000B50363148303820-2020-08-22T15-29-58.bin", 'rb') as f:
    input_data = f.read()
    
short_array = struct.unpack('>' + 'H' * (len(input_data) // 2), input_data)
short_array = short_array - np.mean(short_array)
```

```python
from scipy import signal as sig

fig, axs = plt.subplots(2,1)

f, Pxx_den = sig.periodogram(short_array, 2500)
axs[0].plot(short_array)
axs[1].plot(f, Pxx_den)
```
